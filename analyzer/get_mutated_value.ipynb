## get_normal_trace

```python
from analyzer.utils import parse_traces
from analyzer.decoder import trace_transaction
from analyzer.detection_rules import find_fn
from web3 import Web3

def find_fn_with_args(parsed, fn_names):
    calls = []
    if 'input' in parsed.keys():
        try:
            if parsed['input'][0].fn_name in fn_names:
                calls.append({
                    'input': parsed['input'],
                    'from': parsed['from'],
                    'to': parsed['to'],
                })
        except:
            pass
    if 'calls' in parsed.keys():
        for call in parsed['calls']:
            tmp = find_fn_with_args(call, fn_names)
            for c in tmp:
                calls.append(c)
    return calls

def call_to_json(calls):
    return {
        'fnName': calls['input'][0],
        'args': Web3.toJSON(calls['input'][1])
    }

def get_normal_traces(tx):
    traces = trace_transaction(tx)['result']
    parsed = parse_traces(traces)
    calls = find_fn_with_args(parsed, ["getReserves", "getExpectedRate", "calc_token_amount", "calc_withdraw_one_coin"])
    return {
        'targetAddress': parsed['to'],
        'targetContractName': parsed['to_name'],
        'mutateTarget': [call_to_json(c) for c in calls]
    }
get_normal_traces('0x8076d54315391ca406fa580dbd2f891bc7901ffaf10207aab59cd4138bdd367f')
```
