```python
def parse_data(line):
    data = line.split(',')
    if len(data) == 5:
        return line
    

    for d in data[5:]:
        data[4] = '{}\,{}'.format(data[4], d)

    
    for d in data[0:5]:
        data[0] = '{},{}'.format(data[0], d)

    return data[0]
```
