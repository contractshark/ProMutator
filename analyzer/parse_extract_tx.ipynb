## pre-processing csv

```python
import os
index = 6
count = 0
log = './tx_list.csv'#.format(index)
f = open(log)
another_log = './formated_tx_list.csv'#.format(index)
mydata = ''
if not os.path.exists(another_log):
    f = open(log, 'r')
    af = open(another_log, 'w')
    datas = f.read().replace('"', "'").split('\n')
    f.close()
    print('total count:', len(datas))
    for data in datas[:]:
#         mydata = mydata +'\n' + parse_data(data)
        parsed = parse_data(data)
        if parsed != '':
            af.writelines(parsed)
        count += 1
        if count % 10000 == 0:
            print(count)
```

```python
def parse_data(line):
    data = line.split(',')
    try:
        data[4] = '"' + data[4]
    except:
        return ''
    data[-1] = data[-1][:-1] + '"\n'
    
    return_data = ''
    for d in data:
        return_data = return_data + "," + d

    return return_data
```
