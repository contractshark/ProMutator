```python
import os
import json
base_path = '/data/contract/analyzer/jsonlog2'
txs = sorted(os.listdir('/data/contract/analyzer/jsonlog2'))
```

```python
def check_sus_dict(sus_dict, rate, suspicious_type):
    if rate in sus_dict.keys():
        if sus_dict[rate]['suspicious']['type'] == suspicious_type:
            return True
    return False

def open_txs_report(txs):
    f = open(os.path.join(base_path, txs), 'r')
    datas = f.readlines()
    sus_rate = {}
    for data in datas:
        parsed = json.loads(data)
        if parsed['suspicious']['type'] == 'value-inconsists' and len(json.loads(parsed['suspicious']['msg'])) == 0:
            continue
#         if parsed['suspicious']['type'] == 'value-inconsists':
        sus_rate[parsed['rate']] = parsed
    
    
    if check_sus_dict(sus_rate, 0.001, 'reverted')  and not check_sus_dict(sus_rate, 1000, 'reverted'):
        print(txs[:-4], sus_rate[0.001]['suspicious']['msg'])
    if check_sus_dict(sus_rate, 1000, 'reverted')  and not check_sus_dict(sus_rate, 0.001, 'reverted'):
        print(txs[:-4], sus_rate[1000]['suspicious']['msg'])

#     if check_sus_dict(sus_rate, 0.99, 'value-inconsists'):
#         if check_sus_dict(sus_rate, 0.95, 'value-inconsists'):
#             if check_sus_dict(sus_rate, 1.01, 'value-inconsists'):
#                 print(txs[:-4])
    
    # heuristic 1
    return sus_rate
    
# dict = open_txs_report(txs[:30])
for tx in txs[:100]:
    sus_rate = open_txs_report(tx)
#     print(sus_rate)
```
